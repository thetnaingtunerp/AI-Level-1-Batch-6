```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles'],
    'Salary': [70000, 85000, 60000, 95000, 80000]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n")

# Filter for rows where Age is greater than 25
filtered_df_age = df[df['Age'] > 25]
print("DataFrame where Age > 25:")
print(filtered_df_age)
print("\n")

# Filter for rows where City is 'New York'
filtered_df_city = df[df['City'] == 'New York']
print("DataFrame where City is 'New York':")
print(filtered_df_city)
print("\n")

# Filter for rows where Age is greater than 25 AND City is 'Los Angeles'
filtered_df_multiple = df[(df['Age'] > 25) & (df['City'] == 'Los Angeles')]
print("DataFrame where Age > 25 AND City is 'Los Angeles':")
print(filtered_df_multiple)
print("\n")

# Filter using .loc for specific conditions and columns
filtered_loc = df.loc[df['Salary'] > 75000, ['Name', 'Salary']]
print("Names and Salaries for individuals with Salary > 75000:")
print(filtered_loc)
```

** Data Correlations**

groupby() ကို သုံးပြီး columns အလိုက် အုပ်စုဖွဲ့နိုင်ပါတယ်။

```python
import pandas as pd

data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [6000, 7000, 4000, 4500, 8000, 7500],
    'Bonus': [500, 600, 200, 250, 800, 750]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Salary average by department
print(df.groupby('Department')['Salary'].mean())

# Sum of salary by department
print(df.groupby('Department')['Salary'].sum())

# Multiple aggregation
print(df.groupby('Department')['Salary'].agg(['mean', 'sum', 'max', 'min']))
```

**Aggregating Multiple Columns**

```python
# Apply aggregation to multiple columns
print(df.groupby('Department')[['Salary', 'Bonus']].mean())

# Different aggregation for each column
print(df.groupby('Department').agg({
    'Salary': 'mean',
    'Bonus': 'sum'
}))
```

**Grouping by Multiple Keys**

```python
# Group by Department + Employee
print(df.groupby(['Department', 'Employee'])['Salary'].sum())
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Region': ['East', 'East', 'West', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Sales': [100, 150, 200, 120, 130, 180, 110, 160],
    'Quantity': [10, 15, 20, 12, 13, 18, 11, 16]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n")

# Create a pivot table to show total sales by Region and Product
pivot_df = pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0)

print("Pivot Table (Total Sales by Region and Product):")
print(pivot_df)
print("\n")

# Another pivot table example: Average Quantity sold per Region and Product
pivot_df_avg_qty = pd.pivot_table(df, values='Quantity', index='Region', columns='Product', aggfunc='mean', fill_value=0)

print("Pivot Table (Average Quantity by Region and Product):")
print(pivot_df_avg_qty)
```

**Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data: Monthly sales for Product A and Product B
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Product A': np.random.randint(100, 200, size=12),
    'Product B': np.random.randint(80, 160, size=12)
}

df = pd.DataFrame(data)
df = df.set_index('Month')
df
```

```python
plt.figure(figsize=(10,8))
df.plot(kind='line', marker='o', color=['green', 'blue'], title='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(['Prod A - Green', 'Prod B - Blue'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.plot(kind='bar', alpha=0.75, rot=45, color=['darkorange', 'darkblue'], title='Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Box Plots:
plt.figure(figsize=(10, 6))
color = {'boxes': 'DarkGreen', 'whiskers': 'DarkOrange', 'medians': 'DarkBlue', 'caps': 'Gray'}
df.plot(kind='box', title='Sales Distribution for Product A and Product B', color=color)
plt.ylabel('Sales')
plt.show()
```

```python
# Histograms:
plt.figure(figsize=(10, 6))
df['Product A'].plot(kind='hist', bins=6, color='purple', edgecolor='black', title='Sales Distribution of Product A')
plt.xlabel('Sales')
plt.show()
```

```python
# Creating Subplots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

df['Product A'].plot(ax=ax[0], color='blue', title='Product A Sales Over Time')
df['Product B'].plot(ax=ax[1], color='green', title='Product B Sales Over Time')

plt.tight_layout()
plt.show()
```

**Extract Data in Date and Time Ranges**

```python
# Load library
import pandas as pd

# Create data frame
df = pd.DataFrame()

# Create datetimes
df['date'] = pd.date_range('1/1/2012', periods = 1000, freq ='H')

print(df.head(5))

# Select observations between two datetimes
x = df[(df['date'] > '2012-1-1 01:00:00') &
       (df['date'] <= '2012-1-1 11:00:00')]

print(x)
```

**Method #2: If the dataset is indexed with time **

```python
# Load library
import pandas as pd

# Create data frame
df = pd.DataFrame()

# Create datetimes
df['date'] = pd.date_range('1/1/2012', periods = 1000, freq ='H')

# Set index
df = df.set_index(df['date'])

print(df.head(5))

# Select observations between two datetimes
x = df.loc['2012-1-1 04:00:00':'2012-1-1 12:00:00']

print(x)
```
