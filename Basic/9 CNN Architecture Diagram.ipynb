```python
# Input Image (e.g., 28x28x1 grayscale image)
#         │
#         ▼
# Convolutional Layer (e.g., 32 filters, 3x3 kernel) + ReLU
#         │
#         ▼
# Pooling Layer (e.g., MaxPooling 2x2)
#         │
#         ▼
# Convolutional Layer (e.g., 64 filters, 3x3 kernel) + ReLU
#         │
#         ▼
# Pooling Layer (MaxPooling 2x2)
#         │
#         ▼
# Flatten Layer (convert 2D feature maps to 1D vector)
#         │
#         ▼
# Fully Connected (Dense) Layer (e.g., 128 neurons) + ReLU
#         │
#         ▼
# Output Layer (Dense, number of classes, e.g., 10) + Softmax
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Build a simple CNN model
model = models.Sequential([
    # Input layer expects 28x28 grayscale images
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 classes for classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Model summary
model.summary()
```


Conv2D layers: 2D convolutional layers with ReLU activation extract spatial features.
MaxPooling2D: Downsamples feature maps to reduce spatial size and computation.
Flatten: Converts 2D feature maps into 1D vector for fully connected layers.
Dense: Fully connected layers for classification.
Output layer uses softmax activation for multi-class classification.

## MNIST Dataset နဲ့ CNN Training, Evaluation, Prediction Example

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# 1. MNIST dataset ကို load လုပ်ခြင်း
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Data Preprocessing
# - Pixel values ကို 0-1 range မှာ normalize လုပ်ခြင်း
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# - CNN input အတွက် shape ကို (28, 28, 1) ပြောင်းခြင်း (grayscale channel)
x_train = np.expand_dims(x_train, axis=-1)  # shape: (num_samples, 28, 28, 1)
x_test = np.expand_dims(x_test, axis=-1)

# 3. CNN Model တည်ဆောက်ခြင်း
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Model ကို compile လုပ်ခြင်း
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5. Model ကို train လုပ်ခြင်း
model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.1)

# 6. Model ကို evaluate လုပ်ခြင်း (Test set accuracy)
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Test accuracy: {test_acc:.4f}')

# 7. Prediction လုပ်ခြင်း
predictions = model.predict(x_test)

# Example: ပထမဆုံး test image ရဲ့ prediction
predicted_label = np.argmax(predictions[0])
print(f'Predicted label for first test image: {predicted_label}')
print(f'Actual label: {y_test[0]}')
```
