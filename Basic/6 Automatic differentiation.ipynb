### Automatic differentiation (AD)

<p style="color:blue;">computational graphs ကိုအသုံးပြုပြီး function တစ်ခုရဲ့ derivatives (gradients) တွေကို အလိုအလျောက် တွက်ချက်ပေးပါတယ်။</p>

```python
# Computational Graph ဆောက်တည်ခြင်း 
import tensorflow as tf

# Variables များဖန်တီးခြင်း
x = tf.Variable(3.0)
y = tf.Variable(2.0)

# Operation များဖန်တီးခြင်း
with tf.GradientTape() as tape:
    z = x**2 + y**3
```

<p>GradientTape ကိုအသုံးပြုခြင်း</p>

```python
# Gradients တွက်ချက်ခြင်း
with tf.GradientTape() as tape:
    z = x**2 + y**3

# z ရဲ့ x နဲ့ y အပေါ် gradients တွက်ချက်ခြင်း
gradients = tape.gradient(z, [x, y])

print(f"dz/dx: {gradients[0]}")
print(f"dz/dy: {gradients[1]}")
```

```python
# Example 1 : Simple Function
# Simple quadratic function
x = tf.Variable(5.0)

with tf.GradientTape() as tape:
    f = x**2 + 3*x + 2

df_dx = tape.gradient(f, x)
print(f"f({x.numpy()}) = {f.numpy()}")
print(f"f'({x.numpy()}) = {df_dx.numpy()}")
```

```python
# Example 2 Neural Network Training
# Simple linear regression
# Data preparation
X = tf.constant([[1.0], [2.0], [3.0], [4.0]])
y = tf.constant([[2.0], [4.0], [6.0], [8.0]])

# Model parameters
W = tf.Variable(tf.random.normal([1, 1]))
b = tf.Variable(tf.zeros([1]))

# Training loop
learning_rate = 0.01
epochs = 100

for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # Forward pass
        y_pred = tf.matmul(X, W) + b
        # Loss calculation (MSE)
        loss = tf.reduce_mean(tf.square(y_pred - y))
    
    # Compute gradients
    gradients = tape.gradient(loss, [W, b])
    
    # Update parameters
    W.assign_sub(learning_rate * gradients[0])
    b.assign_sub(learning_rate * gradients[1])
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.numpy()}")
```

```python
# Training Neural Network Using TensorFlow's AD Capabilities
import tensorflow as tf
# Generate some random data
X = tf.random.normal((100, 1))
y = 3*X + tf.random.normal((100, 1))

# Define a simple neural network
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='sgd', loss='mse')

# Train the model using AD for gradient computation
model.fit(X, y, epochs=10, verbose=0)

# Print the trained weights
print("Trained weights:", model.layers[0].get_weights()[0])
```

<p>Comparison of Manual Gradient Computation vs TensorFlow AD</p>

```python
import tensorflow as tf

# Define a simple mathematical function
def func(x):
    return x**2 + 5*x + 3

# Define the input variable
x = tf.Variable(2.0)

# Manual gradient computation
with tf.GradientTape() as tape:
    y = func(x)

manual_grad = 2*x + 5

# Automatic differentiation
with tf.GradientTape() as tape:
    y = func(x)

auto_grad = tape.gradient(y, x)

print("Manual Gradient:", manual_grad.numpy())
print("Automatic Gradient:", auto_grad.numpy())
```

<p>Advanced Automatic Differentiation using Custom Gradients:</p>

```python
import tensorflow as tf

# Define a function for which you want to compute gradients
def custom_function(x):
    return tf.square(x)

# Define the gradient of the custom function
@tf.custom_gradient
def custom_function_with_gradient(x):
    # Forward pass
    y = custom_function(x)

    # Define gradient function
    def grad(dy):
        return 2 * x * dy  # Gradient of x^2 is 2x

    return y, grad

# Example usage
x = tf.constant(3.0)
with tf.GradientTape() as tape:
    tape.watch(x)
    y = custom_function_with_gradient(x)

# Compute gradients
grad = tape.gradient(y, x)

print("Function output:", y.numpy())
print("Gradient:", grad.numpy())
```
