<h2 data-start="88" data-end="122">🌱 Gradient Boosting ဆိုတာဘာလဲ?</h2>
<p data-start="123" data-end="384">Gradient Boosting (GB) က <strong data-start="148" data-end="169">Ensemble Learning</strong> technique တစ်ခုပါ။<br data-start="188" data-end="191">
တစ်ခါတည်း decision tree တစ်ပင်နဲ့မဟုတ်ပဲ၊ <strong data-start="233" data-end="289">အများကြီး weak learners (ပုံမှန် decision tree များ)</strong> တစ်ခုပြီးတစ်ခု ဆောက်ပြီး <strong data-start="315" data-end="365">အကျိုးစာရင်းကို အဆင့်ဆင့်တိုးတက်အောင် လေ့ကျင့်</strong> သွားတဲ့နည်းလမ်းပါ။</p>
<p data-start="386" data-end="404">➡️ <strong data-start="389" data-end="402">Main Idea</strong></p>
<ol data-start="405" data-end="811">
<li data-start="405" data-end="485">
<p data-start="408" data-end="485">ပထမဆုံး <strong data-start="416" data-end="440">အလွယ်ဆုံး prediction</strong> (baseline) တစ်ခုလုပ်တယ် (ဥပမာ mean value).</p>
</li>
<li data-start="486" data-end="532">
<p data-start="489" data-end="532">Model က <strong data-start="497" data-end="518">error (residuals)</strong> ကိုတွက်တယ်။</p>
</li>
<li data-start="533" data-end="644">
<p data-start="536" data-end="644">Error ကို သက်ဆိုင်ရာ Gradient (Loss function ၏ derivative) နဲ့သုံးပြီး <strong data-start="607" data-end="641">အသစ် decision tree လေ့ကျင့်တယ်</strong>။</p>
</li>
<li data-start="645" data-end="698">
<p data-start="648" data-end="698">အသစ်ရတဲ့ tree ကို <strong data-start="666" data-end="687">ပုံစံထဲ ထပ်ပေါင်း</strong> ထည့်တယ်။</p>
</li>
<li data-start="699" data-end="811">
<p data-start="702" data-end="811">ဒီလိုနဲ့ လိုအပ်သလောက် weak learners (trees) တွေထပ်ပေါင်းပြီး <strong data-start="763" data-end="790">တိုးတက်တဲ့ strong model</strong> တစ်ခု ရအောင်လုပ်တယ်။</p>
</li>
</ol>

<h2 data-start="818" data-end="877">🧮 Gradient Boosting အလုပ်လုပ်ပုံ (Example – Regression)</h2>
<ol data-start="878" data-end="1204">
<li data-start="878" data-end="911">
<p data-start="881" data-end="911">Prediction ပထမဆုံး = mean(y)</p>
</li>
<li data-start="912" data-end="942">
<p data-start="915" data-end="942">Residual = y - prediction</p>
</li>
<li data-start="943" data-end="1003">

</li>
<li data-start="1004" data-end="1167">

<h2 data-start="1211" data-end="1260">📊 Gradient Boosting Parameters (Scikit-Learn)</h2>
<p data-start="1261" data-end="1358">Scikit-Learn မှာ <code data-start="1278" data-end="1306">GradientBoostingClassifier</code> နဲ့ <code data-start="1311" data-end="1338">GradientBoostingRegressor</code> ကို သုံးနိုင်ပါတယ်။</p>
<p data-start="1360" data-end="1380"><strong data-start="1360" data-end="1378">Key Parameters</strong></p>
<ul data-start="1381" data-end="1694">
<li data-start="1381" data-end="1429">
<p data-start="1383" data-end="1429"><code data-start="1383" data-end="1397">n_estimators</code> → Tree အရေအတွက် (default=100)</p>
</li>
<li data-start="1430" data-end="1508">
<p data-start="1432" data-end="1508"><code data-start="1432" data-end="1447">learning_rate</code> → Each tree အထောက်အပံ့ အမြောက်အများကို ထိန်း (default=0.1)</p>
</li>
<li data-start="1509" data-end="1544">
<p data-start="1511" data-end="1544"><code data-start="1511" data-end="1522">max_depth</code> → Trees တွေရဲ့ အနက်</p>
</li>
<li data-start="1545" data-end="1628">
<p data-start="1547" data-end="1628"><code data-start="1547" data-end="1558">subsample</code> → Dataset fraction to train each tree (default=1.0 → whole dataset)</p>
</li>
<li data-start="1629" data-end="1694">
<p data-start="1631" data-end="1694"><code data-start="1631" data-end="1637">loss</code> → Loss function (deviance, exponential, squared_error …)</p>
</li>
</ul>
