```python
# Task 1: Question Answer
print("Random initialization is useful because it breaks symmetry, ensuring that neurons learn different features. If all weights started at the same value, all neurons would update identically, limiting the model's learning capacity.")


# Task 2 & 3: Create train and validation transforms
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm

train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])


# Task 4: Create val_loader
dataset_path = './images_dataSAT/'
full_dataset = torchvision.datasets.ImageFolder(root=dataset_path, transform=train_transform)

train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])

val_dataset.dataset.transform = val_transform
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)


# Task 5-8: Questions
print("\nQ5: tqdm is used for displaying progress bars during loops (training epochs, batches).")
print("Q6: train_loss, train_correct, train_total are reset each epoch to compute metrics for that specific epoch only.")
print("Q7: torch.no_grad() is used in validation to disable gradient computation, saving memory and speeding up computation since gradients aren't needed for evaluation.")
print("Q8: Two metrics: Accuracy and F1-Score (or Precision/Recall).")


# Task 9: Plot Model Loss (assuming training history exists)
# After training a model, you'd have history
# plt.plot(train_losses, label='Training Loss')
# plt.plot(val_losses, label='Validation Loss')


# Task 10: Get predictions and labels from val_loader
def get_predictions_and_labels(model, val_loader, device='cpu'):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())
    return all_preds, all_labels

# Assuming model is defined and trained
# all_preds, all_labels = get_predictions_and_labels(model, val_loader)
```
