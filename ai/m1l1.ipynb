```python
# Task 1: Determine the shape of a single image
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# Assuming image_data is provided
# print(f"Image shape: {image_data.shape}")


# Task 2: Display first four images from class_0_non_agri
non_agri_dir = './images_dataSAT/class_0_non_agri/'
non_agri_images = [os.path.join(non_agri_dir, f) for f in os.listdir(non_agri_dir) 
                   if f.endswith(('.png', '.jpg', '.jpeg'))][:4]

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for i, img_path in enumerate(non_agri_images):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(f'Image {i+1}')
plt.show()


# Task 3: Create sorted agri_images_paths list
dir_agri = './images_dataSAT/class_1_agri/'
agri_images_paths = sorted([os.path.join(dir_agri, f) for f in os.listdir(dir_agri)
                            if f.endswith(('.png', '.jpg', '.jpeg'))])


# Task 4: Count agricultural land images
agri_count = len(agri_images_paths)
print(f"Number of agricultural images: {agri_count}")


# Task 5: Display first four agricultural images
fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for i, img_path in enumerate(agri_images_paths[:4]):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(f'Agri {i+1}')
plt.show()
```
