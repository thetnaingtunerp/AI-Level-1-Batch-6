```python
# Task 1: Create transformation pipeline
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt

custom_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    transforms.RandomRotation(degrees=45),
    transforms.ToTensor(),
])


# Task 2: Load dataset using ImageFolder
dataset_path = './images_dataSAT/'
imagefolder_dataset = torchvision.datasets.ImageFolder(
    root=dataset_path,
    transform=custom_transform
)


# Task 3: Print class names and indices
print("Class names and indices:")
for class_name, class_idx in imagefolder_dataset.class_to_idx.items():
    print(f"  {class_name}: {class_idx}")


# Task 4: Retrieve a batch and display shapes
imagefolder_loader = DataLoader(imagefolder_dataset, batch_size=8, shuffle=True)
images, labels = next(iter(imagefolder_loader))
print(f"Batch images shape: {images.shape}")
print(f"Batch labels shape: {labels.shape}")


# Task 5: Display images in the batch
def imshow(img_batch):
    img_batch = img_batch.numpy().transpose((0, 2, 3, 1))
    fig, axes = plt.subplots(2, 4, figsize=(12, 6))
    axes = axes.flatten()
    for i in range(min(8, len(img_batch))):
        axes[i].imshow(img_batch[i])
        axes[i].axis('off')
    plt.show()

imshow(images)
```
